```python
import pandas as pd
import numpy as np
  
import os, sys, subprocess



DATABASE = "202103-EXP.mdb" # database name 

# Get table names using mdb-tables
table_names = subprocess.Popen(['mdb-tables', '-1', DATABASE], stdout=subprocess.PIPE).communicate()[0]

table_names=table_names.decode("utf-8") 
tables = table_names.split('\n')[:-1]

print(tables)
total_csv=[]
for table in tables:
    if table != '':
        filename = table.replace(' ','_') + '.csv'
        
        filename="finallll"+filename
        total_csv.append(filename)
        print('Exporting ' + table)
        with open(filename, 'wb') as f:
            subprocess.check_call(['mdb-export', DATABASE, table], stdout=f)

        
```

```python
count=0
all_xls_file=[]
for filename in total_csv:

    for chunk in pd.read_csv(filename, chunksize=10000):
       # importing the module

       #saving xlsx file
        
        name=table+str(count)+'.xlsx'
        print(name)
        all_xls_file.append(name)
        to_xls = pd.ExcelWriter(table+str(count)+'.xlsx')
        chunk.to_excel(to_xls, index = False)

        to_xls.save()
        count=count+1
```
